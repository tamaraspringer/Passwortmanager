```python
import sys
import time


class Krankheit():
    def __init__(self,definition):
        self.definition = definition
        
    def set_definition(self, definition):
        self.definition = definition
    def get_definition(self):
        return str(self.definition)     
                   
class Frage():
    def __init__(self,text):
        self.text = text
        
    def set_text(self, text):
        self.text = text    
    def getPrint(self):
        return  print(self.text)   
    
    def __str__(self):
        return self.text  
       
class Schizophrenie(Krankheit):
    def __init__(self,definition, isKrank, prozentsatz):
        super().__init__(definition) 
        self.isKrank= isKrank  
        self.prozentsatz = prozentsatz 
              
    def set_isKrank(self, isKrank):
        self.isKrank = isKrank
    def get_isKrank(self):
        return str(self.isKrank) 
    
    def set_prozentsatz(self, prozentsatz):
        self.prozentsatz = prozentsatz
    def get_prozentsatz(self):
        return str(self.prozentsatz) 
    
    def printDefinition(self):
        return  print(self.definition)
    
    def __str__(self):
        if self.isKrank == False:
            return "Hey, du hast keine Schizophrenie"
        if self.isKrank == True:
            return  "Du bist " + str(self.prozentsatz) + " % " + "schizophren."
     
    
def printDelay(range, delay=0.100):
   for a in range:
    sys.stdout.write(a)
    sys.stdout.flush()
    time.sleep(delay) 
    
def startWieder( antwort,frage):
    print("Ich check es nicht, versuch mal wieder es zu beantworten (ja oder nein)")
    antwort = str(input(frage))
    print("dein Antwort ist " + antwort)
    return antwort
     
schizophrenie = Schizophrenie(" ", False, 0)    
schizophrenie.definition = "Schizophrenie ist eine psychische Störung, bei der die Gedanken und Wahrnehmungen der Betroffenen verändert sind"
frage1 = Frage("Weißt du, was Schizophrenie ist? ")
frage2 = Frage("Willst du dich testen, wie schizophren bist du?")
frage3 = Frage("Ihre Ausdauer und Motivation in Schule, Ausbildung oder Arbeit und bei Freizeitunternehmungen hat auffällig nachgelassen. Stimmt es?")
frage4 = Frage("Hast du einen imaginären Freund?")
frage5 = Frage("Hast du einen Psychiater?")
frage6 = Frage("Denkst du, du bist schizophren?")


antwort1 = str(input(frage1)) 
if antwort1 == "nein":
    print("Dann versuche ich es zu erklären")
    printDelay("**********************\n")
    schizophrenie.printDefinition()
elif   antwort1 == "ja": 
    print("Es klingt schon nicht schlecht") 
else :
     startWieder(antwort1, frage1)
    
antwort2 = str(input(frage2))             
if antwort2 == "nein":
    sys.exit("Es klingt schlecht, aber wie du willst")
elif  antwort2 == "ja": 
    print("Lass uns anfangen") 
else : 
    startWieder(antwort2, frage2)
    
frage3.getPrint()      
antwort3 = str(input(frage3)) 
if antwort3 == "nein" or antwort3 == "ja" :
  print("dein Antwort ist " + antwort3)
else : 
   antwort3 = startWieder(antwort3, frage3)
    

frage4.getPrint() 
antwort4 = str(input(frage4))   
if antwort4 == "nein" or antwort4 == "ja" :
  print("dein Antwort ist " + antwort4)
else : 
   antwort4 = startWieder(antwort4, frage4)

frage5.getPrint() 
antwort5 = str(input(frage5))    
if antwort5 == "nein" or antwort5 == "ja" :
  print("dein Antwort ist " + antwort5)
else : 
   antwort5 = startWieder(antwort5, frage5)

frage6.getPrint()   
antwort6 = str(input(frage6))  
if antwort6 == "nein" or antwort6 == "ja" :
  print("dein Antwort ist " + antwort6)
else : 
   antwort6 = startWieder(antwort6, frage6)   

antwortList = (antwort3,antwort4,antwort5,antwort6)
antwortJaCount = antwortList.count("ja")

if antwortJaCount <=0:
    schizophrenie.isKrank = False
elif   antwortJaCount == 1: 
     schizophrenie = Schizophrenie(" ",True,25)
elif   antwortJaCount == 2: 
      schizophrenie = Schizophrenie(" ",True,50)
elif   antwortJaCount == 3: 
      schizophrenie = Schizophrenie(" ",True,75)
elif   antwortJaCount == 4:  
     schizophrenie = Schizophrenie(" ",True,100)   
print(schizophrenie)          
      
```
